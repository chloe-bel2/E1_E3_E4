```python
import pandas as pd
```

```python
df_pe = pd.read_csv("C:\\Users\\Utilisateur\\Documents\\Prepa_Diplome\\Projet_CO\\DB_pco\\10k_offres.csv")
df_nantes_metro = pd.read_csv("C:\\Users\\Utilisateur\\Documents\\Prepa_Diplome\\Projet_CO\\DB_pco\\nantes_metropole.csv")
```

# Agrégation des dataframes offres d'emploi

```python
df_pe.head()
```

```python
df_nantes_metro.head()
```

```python
df_nantes_metro = df_nantes_metro[['ref_offre', 'descri_mission']]
```

```python
df_nantes_metro
```

```python
df_pe.columns = df_nantes_metro.columns

# Concaténation des deux DataFrames
df_concat = pd.concat([df_pe, df_nantes_metro], axis=0)
df_concat
```

```python
df_concat.isna().sum()
```

# Importation du dataframe dans la base SQL "db_pco"

```python
from sqlalchemy import create_engine

username = 'postgres'
password = 'postgre'
host = 'localhost'
port = '5432'      
database = 'db_pco'

# URI de connexion
connection_string = f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}"

# Objet engine SQLAlchemy : gère la connexion automatiquement (pas besoin de la fermer)
engine = create_engine(connection_string)

# Try / Except pour s'assurer 
try:
    df_concat.to_sql('offres_emploi', engine, if_exists='replace', index=False)
    print("Les données ont été importées avec succès dans PostgreSQL.")
except Exception as e:
    print(f"Erreur lors de l'importation des données : {e}")
```
